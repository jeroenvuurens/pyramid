```python
import cv2

cap = cv2.VideoCapture(0)  # usually 0 for first USB cam
ret, frame = cap.read()
if ret:
    cv2.imwrite("usb_capture.jpg", frame)
cap.release()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("capture.jpg")
img = cv2.resize(img, (640, 360))  # downscale for display
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

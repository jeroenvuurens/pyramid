```python
# Imports
import math
import socket
import json
import time
```

```python
# Utility function to rotate a point
def rotate_point(x, y, cx, cy, angle):
    rad = math.radians(angle)
    cos_a = math.cos(rad)
    sin_a = math.sin(rad)
    dx = x - cx
    dy = y - cy
    rx = cx + cos_a * dx - sin_a * dy
    ry = cy + sin_a * dx + cos_a * dy
    return [rx, ry]
```

```python
# Function to update and send only the moving object as a rotated polygon
def update_obj1_rotated():
    radius = 200
    center_x, center_y = 500, 400
    rect_w, rect_h = 400, 300
    for angle in range(0, 360, 10):
        cx = center_x + radius * math.cos(math.radians(angle))
        cy = center_y + radius * math.sin(math.radians(angle))
        corners = [
            [cx-rect_w//2, cy-rect_h//2],
            [cx+rect_w//2, cy-rect_h//2],
            [cx+rect_w//2, cy+rect_h//2],
            [cx-rect_w//2, cy+rect_h//2]
        ]
        rotated = [rotate_point(x, y, cx, cy, angle) for x, y in corners]
        obj1 = {
            'name': 'obj1',
            'structure': 'rotated_rect_with_circle',
            'params': {
                'rect_corners': [[int(x), int(y)] for x, y in rotated],
                'inner_circle': [int(cx), int(cy), 75]
            }
        }
        cmd = {
            'structures': {
                'rotated_rect_with_circle': {
                    'components': [
                        {'name': 'rect_corners', 'type': 'polygon', 'style': 'red_thick'},
                        {'name': 'inner_circle', 'type': 'circle', 'style': 'green_thin'}
                    ]
                }
            },
            'objects': [obj1]
        }
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.connect(('127.0.0.1', 9999))
        s.send(json.dumps(cmd).encode())
        s.close()
        time.sleep(0.05)
```

```python
# Run the animation
update_obj1_rotated()
```
